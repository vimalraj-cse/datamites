```python
col=['Gender','Married','Dependents','Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History']
```

```python
for column in col:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0],inplace=True)
    else:
        df[column].fillna(df[column].median(),inplace=True)
df
        
```

```python
df.info()
```

```python
df.drop('Loan_ID',axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['Dependents'].value_counts()
```

```python
df['Dependents'].replace('3+',3,inplace=True)
```

```python
df['Dependents']=df['Dependents'].astype('int')
```

```python
df.Dependents.value_counts()
```

```python
col2=df.select_dtypes(include=['object']).columns
col2
```

```python
from sklearn.preprocessing import *
lb=LabelEncoder()
for i in col2:
    df[i]=lb.fit_transform(df[i])
df
```

```python
from sklearn.preprocessing import *
ms=MinMaxScaler()
scaled_value=ms.fit_transform(df.iloc[:,:-1])
x=pd.DataFrame(scaled_value,columns=df.columns[:-1])
x
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import *
```

```python
model=SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
params={'C':[0.1,1,10,100],
        'gamma':[1,0.1,0.01,0.001]
       ,'kernel':['linear','rbf','poly']}
grid=GridSearchCV(SVC(),params,refit=True,verbose=2,n_jobs=-1)
grid.fit(x_train,y_train)
print(f'Best_Parameters:{grid.best_params_}')
```

```python
print('Best_Parameters:',grid.best_params_)
```

```python
best_params= {'C': 0.1, 'gamma': 1, 'kernel': 'poly'}
grid=SVC(**best_params)
grid.fit(x_train,y_train)
y_pred=grid.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
